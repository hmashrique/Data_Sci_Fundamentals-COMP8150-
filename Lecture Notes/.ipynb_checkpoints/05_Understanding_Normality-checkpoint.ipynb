# Understanding Normality

+ mathematical notion versus real life

Mathematical notion of normal:
+ There's a central tendency (mean/average).
+ The distribution is symmetric
+ Things gradually deviate from the central tendency.

---
### Shapiro test

Test (null) hypothesis: the data is drawn from a normally distributed population.

After we run the statistical test, the procedure returns a p-value.

The p-value: the probability of observing the data, suppose that the test hypothesis is true.

Two cases:
* p-value is very low: reject test hypothesis.
* High p-value: not enough evidence to reject the test hypothesis.

The p-value does not indicate the magnitude of the difference between the groups, nor does it measure the importance of the difference. It merely quantifies how likely it is to observe the data if the null hypothesis were true. 



```python
import warnings
warnings.filterwarnings('ignore')
```

```python
_, p_value = shapiro(iris['SepalLength'])
p_value
```

This means that assuming sepal length is normally distributed, the probability that we observe `iris['SepalLength']` is 0.01018.


Intuitively, because this probability is very small, we say that it's unlikely that sepal length is normally distributed.

Mathematically, we need to be systematic. What people do is comparing the p-value to a commonly agreed upon threshold, e.g. 0.05.

If the p-value is less than 0.05, we conventionally reject the test hypothesis.

However, if we want to strictly threshold to reject the test hypothesis, we can set the threshold to be 0.01.  In this case, the p-value is larger than 0.01.  We cannot conclude that sepal length is not normally distributed.

#### Compare the distribution of sepal lengths to the (mathematically) normal distribution

Which normal distribution? The one with the same mean and standard deviation.

```python
from numpy.random import normal
from seaborn import displot

normal_sepals = normal(
    iris['SepalLength'].mean(), 
    iris['SepalLength'].std(),
    1000)
displot(normal_sepals, kind='kde', height=2)
```

```python
displot(data=iris, x='SepalLength', kind='kde', height=2)
```

### QQ plot

A QQ plot compares visually the distribution of the normal distribution with same mean and standard deviation with the distribution of the observed data.

```python
from scipy.stats import shapiro
from statsmodels.api import qqplot

def test_for_normality(data):
    shapiro_test = shapiro(data)
    fig = qqplot(data, line='s')
    return round(shapiro_test.pvalue,4)
```

```python
test_for_normality(iris['SepalLength'])
```

```python
test_for_normality(normal_sepals)
```
