# Principal Component Analysis (PCA)

Preliminaries:
+ Orthogonal/uncorrelated features 
+ Information content: self information, entropy
+ Variance - captures variety and information content.
    * Normal distributions: differential entropy = ln( std * sqrt(2epi) )
+ Variance explained
+ Correlation

The main goal of PCA is to reduce the dimensionality of a dataset by transforming the original variables into a new set of variables, called principal components, which are uncorrelated/orthogonal and ordered such that the first few retain most of the variation present in all of the original variables.

### Principal Component Analysis (PCA)

PCA transform a set of features into new dimensions (features), by "rotating" the original features.

The new features are called "principal components".

The first principal component has the highest variance.

The second principal component has the second highest variance.

and so on.

All the principal components are mutually orthogonal.

```python
pca = PCA()
iris_features = iris.drop(columns=['Species'])

iris_pca = pandas.DataFrame(pca.fit_transform(iris_features))
iris_pca.var()
```

```python
iris_pca.sample()
```

```python
seaborn.relplot(data=iris_pca, x=0, y=1, hue=iris['Species'], height=3)
```

```python
abalone_features = abalone.drop(columns=['Sex'])

pca = PCA()
abalone_pca = pandas.DataFrame(
    pca.fit_transform( StandardScaler().fit_transform(abalone_features) )
)
abalone_pca.sample()
```

```python
seaborn.relplot(data=abalone_features, x='Length', y='Diameter', hue=abalone['Rings'], height=2)
```

```python
seaborn.relplot(data=abalone_pca, x=0, y=1, height=2, hue=abalone['Rings'])
```

Which original features make up the principal components, e.g. the first one?

Answer: all of them. But some contribute more than the others.
