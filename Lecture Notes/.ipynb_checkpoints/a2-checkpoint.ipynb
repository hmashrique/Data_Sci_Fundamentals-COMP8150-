## Problem 1

Does smoking status appear to influence tipping (tip percentage) behavior?

Do the comparison numerically and visually.  For visual comparison, use both catplot and displot.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python

# Import the tips dataset into `tips`.

tips= pd.read_csv("../Datasets/tips.csv")

tips.sample(3)
```

```python
# Create a column `tip_percentage` using `tip` and `total_bill`.

tips['tip_percentage']= ((tips['tip']/tips['total_bill'])*100).round(1)
```

```python
tips.sample(3)
```

```python
tips['smoker'].value_counts()
```

We can see that non-smokers are more than smokers in the tips dataset. 
Lets plot a categorical plot to observe how much smokers and non-smokers tip against their bill.

```python
sns.catplot(data=tips, y='tip_percentage', x='smoker', kind='bar', height=2.5)

plt.show()
```

So, it looks like smokers tip a bit more than non-smokers. lets see their mean tip percentages as a group.

```python
tips.groupby('smoker')[['tip_percentage']].mean()
```

The mean tip percentage of the 2 groups is very close. Lets look at the distribution(by using distplot) of the 2 groups to understand their tipping pattern.

```python

sns.displot(data=tips, x='tip_percentage', hue='smoker', kind='kde', height=3)

plt.show()
```

Similarly, as the catplot showed a very close tipping percentage between the 2 groups, looking at the distribtuion of the tip percentage for the groups also show that smokers and non-smokers demonstrate a similar trend of tipping. 

Both the groups tip percentage is mostly around the 5-30 percent region with the main difference being their density and shape due to the number of data points in each group.

## Problem 3

Determine if the tip percentage is distributed normally. Include in your analysis (i) a visualization the distribution of the data, (ii) a numerical statistical test, and (iii) visual statistical test.


(i) Lets plot a displot (kde plot) to see the distribution of `tip_percentage`.

```python
## (i) visually displaying the distribution:

sns.displot(data=tips, x='tip_percentage', kind='kde', height=3)

plt.show()
```

The distribution looks like close to a bell curve assuming a normal distribution. Let use a qq plot to see visually if the data is actually normally distrubuted.

```python
# (iii) Visual statistical test:

from statsmodels.api import qqplot

data= tips['tip_percentage']

fig = qqplot(data, line='s')

plt.show()
```

In the qq plot, if the data was normally distributed, the data point would align with the red line perfectly. We can see that data points at the start and also in the end are not aligned to the line. So its not certain if the data is actually normally distributed. Lets use a numerical statistical test(shapiro test) to confirm its normality.

```python
# (ii) Numerical statistical test (finding the p value:)

from scipy.stats import shapiro

shapiro_test = shapiro(data)

print(f"p-value= {shapiro_test.pvalue}")
```

We can see that the p-value for `tip percentage` is very low(<0.05). This confirms that the null hypothesis that assumes "the data is normally distributed" is rejected. So, `tip percentage` is not normally distributed.

## Problem 4

Determine if the tip percentage is distributed normally during lunch time, and dinner time. Include in your analysis (i) a visualization the distribution of the data, (ii) a numerical statistical test, and (iii) visual statistical test.


```python
tips.sample(4)
```

Lets use a displot to see the distribution for `lunch` and `dinner` time `tip perentage`. 

```python
## Visualize lunch and dinner tip percentage distribution

sns.displot(data=tips, x='tip_percentage', hue='time', kind='kde', height=3)

plt.show()
```

The distribution for `lunch` and `dinner` time `tip percentage` looks similar in shape but the `lunch` time distribution looks more of a normal distribution than `dinner`. Lets use a qq plot to see if they are normally distributed.

```python
# (ii) Visual statistical test

# Divide the dinner and lunch time data from tips data.

dinner_data= tips[tips['time']=='Dinner']
lunch_data= tips[tips['time']=='Lunch']

```

```python
# plot qq plot for Dinner time data

from statsmodels.api import qqplot

fig = qqplot(dinner_data['tip_percentage'], line='s')

plt.title('Tip percentage during Dinner')
plt.show()
```

Looking at the qq plot for `dinner` data, it looks like lots of data points are not aligned with the red line, suggesting dinner data not being normally distributed. 

```python
# plot qq plot for Lunch time data

fig2 = qqplot(lunch_data['tip_percentage'], line='s')

plt.title('Tip percentage during Lunch')
plt.show()
```

Looking at the qq plot for `lunch` data, it looks like most of the data points are aligned with the red line, suggesting `lunch` data being normally distributed. 

Now, lets use the shapiro test to confirm the normality of both the `lunch` and `dinner` data.

```python
## finding p values for lunch and dinner:

lunch_tip= lunch_data['tip_percentage']

dinner_tip= dinner_data['tip_percentage']


shapiro_lunch = shapiro(lunch_tip)
print(f"p-value for lunch tip = {shapiro_lunch.pvalue}")

shapiro_dinner = shapiro(dinner_tip)
print(f"p-value for dinner tip = {shapiro_dinner.pvalue}")

```

So the p-value confirms that,
- `Lunch` tip percentage is normally distributed(>0.05).
- `Dinner` tip percentage is not normally distributed(<0.05).

## Problem 5

Is there a difference in people's spending habit between weekends and weekdays?

Is this difference statistically significant?

Do your analyses numerically and visually. Explain why your choice of statistical test of difference.




```python
tips.sample(4)
```

```python
tips['day'].unique()
```

To find spending habit in `weekends` and `weekdays`, lets first label our data points into `weekday` and `weekend` for ease of calculation.

```python
# define function to calculate weekend and weekdays

def day_categorize(day):
    #weekdays=['Mon','Tues','Wed','Thur','Fri']
    weekends= ['Sat','Sun']
    
    if day in weekends:
        day_type='weekend'  
    else:    
        day_type='weekday'
    
    return day_type
    
tips['day_type']= tips['day'].apply(day_categorize)

tips['day_type'].value_counts()
```

```python
tips.sample(5)
```

```python
tips.groupby('day_type')[['tip_percentage']].mean()
```

```python
sns.catplot(data=tips, y='tip_percentage', x='day_type', kind='bar', height=2.5)

plt.show()
```

We can see that the mean `tip percentage` for `weekends` and `weekdays` are very close(both numerically and visually). Lets have a closer look to see the trend of tipping in those times by plotting a displot.

```python

# Set color palette for each category
palette = {'weekday': 'black', 'weekend': 'red'}

sns.displot(data=tips, x='tip_percentage', kind='hist', hue='day_type', color=['red','green'], height=4, palette= palette)

plt.show()
```

Looking at the distribution, the shape of both the groups look similar. Its hard to tell if the tipping behavior difference is statistically different. Lets see the p-value of these times and decide on the test to understand their significance.

```python
# separate weekend and weekday tips data

weekend_tip = tips[tips['day_type']=='weekend']
weekday_tip = tips[tips['day_type']=='weekday']
```

```python
# Do a shapiro test for both to test for normality.

shapiro_weekend = shapiro(weekend_tip['tip_percentage'])
print(f"p-value for weekend tip = {shapiro_weekend.pvalue}")

shapiro_weekday = shapiro(weekday_tip['tip_percentage'])
print(f"p-value for weekday tip = {shapiro_weekday.pvalue}")
```

So, here the p-value of `weekend` tip percentage is very small(< 0.05), which means the data is not normally distributed. In this scenario, we need to do a non parametic test to test for statistical significance. To test for significance, we need to use the `mannwhitneyU` test.

```python
from scipy.stats import mannwhitneyu
# from scipy.stats import ttest_ind


# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(weekend_tip['tip_percentage'], weekday_tip['tip_percentage'])

print("Mann-Whitney U statistic:", statistic)
print("P-value:", p_value)
```

We can see the p-value for the weekend and weekday tip percentage is around 0.303 (> 0.05), which suggests that the difference in tipping is not statistically significant for `weekends` and `weekdays`.
