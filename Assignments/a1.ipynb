**Problem 9**

Define a custom function to calculate the range (difference between max and min) of 'mpg' values for each 'cylinders' group.

Use this custom function in a `groupby` operation to calculate the 'mpg' range for each 'cylinders' group and display the results.

Display the results, and then analyze whether vehicles with more cylinders tend to have a reduced range of 'mpg' values compared to those with fewer cylinders.

**Solution 9**

1. Function `cylinder_range()` is defined.
2. Using the function, groupby `autompg` data by `cylinders` and find the `max` and `min` of the `mpg` column data. Store the result in `aggdata`.
3. Using  `aggdata`, find the difference of max and min of `mpg` of each `cylinder` type. Store it in `aggdata` creating a new `range` column. Return the `aggdata` as final result.  

```python
def cylinder_range():
    
    aggdata= autompg.groupby('cylinders')[['mpg']].agg(['max','min'])
    
    range= aggdata['mpg']['max'] - aggdata['mpg']['min']
    
    aggdata['range']= range
    
    return aggdata

cylinder_range()
```

Looking at the range for each cylinder type, as the cylinders increased, the range of 'mpg' usually increased with it. But for lower cylinder cars, the range was lot smaller.

**Problem 10**

Categorizes vehicles into 'Light', 'Medium', and 'Heavy' based on 'weight_kg'. Use the following criteria: 'Light' (< 1500 kg), 'Medium' (1500-2000 kg), 'Heavy' (> 2000 kg). Use the `apply` method to create a new column 'weight_category' based on this custom function.

Analyze the results to determine the relationship between vehicle weight/weight_category and fuel efficiency.


**Solution 10**

1. Define custiom function `vehicleCat()`. 
2. Inside the function, define the weight categories 'Light','Medium' and 'Heavy' based on the `weight_kg` column.
3. Create a new column `weight_category` on autompg data and use the `apply()` method on the `vehicleCat` function.
4. Display the new `autompg` data.

```python
def vehicleCat(weight):  # custom function
    if weight<1500:
        size='Light'
    elif weight>2000:
        size='Heavy'
    else:
        size='Medium'
    
    return size

autompg['weight_category']= autompg['weight_kg'].apply(vehicleCat)  # creating new column and applying the custom function vehicleCat.

autompg.sample(5)

        
```

To find the relationship between vehicle `weight/weight_category` and `fuel efficiency`, lets find the mean `mpg` of each weight category.

```python
autompg.groupby('weight_category')[['mpg']].mean()
```

Looking at the mean mpg for each weight category, we can say that as the weight gets heavier, the fuel efficiency decreases for cars.
